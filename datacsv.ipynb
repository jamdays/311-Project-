```python
def quantify_hot_sauce(frame):
    frame.iloc[:, 8] = frame.iloc[:, 8].apply(lambda x: 0 if 'None' in str(x).lower() 
                                    else 1 if 'mild' in str(x).lower() 
                                    else 2 if 'medium' in str(x).lower() 
                                    else 3 if 'hot' in str(x).lower()
                                    else 4 )

quantify_hot_sauce(df)
```
