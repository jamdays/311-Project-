```python
import pandas as pd
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('data/cleaned_data_combined_modified.csv')

# Set the plot style
sns.set(style="whitegrid")

# Create a KDE plot
plt.figure(figsize=(10, 6))
sns.kdeplot(
    data=df,
    x="Q1: From a scale 1 to 5, how complex is it to make this food? (Where 1 is the most simple, and 5 is the most complex)",
    hue="Label",  # Group by food type
    fill=True,  # Fill the area under the curve
    common_norm=False,  # Normalize each group independently
    alpha=0.5  # Transparency
)

# Add labels and title
plt.xlabel("Complexity (1 = Simple, 5 = Complex)")
plt.ylabel("Density")
plt.title("Smoothed Distribution of Complexity Ratings by Food Type")
plt.legend(title="Food Type")
plt.show()
```

```python
df = pd.read_csv('data/clean_data.csv')
t = pd.read_csv('data/cleaned_data_combined_modified.csv')['Label']

df = pd.concat([df, t], axis=1)

# Melt the dataframe to make it suitable for grouped bar plotting
melted_df = df.melt(
    id_vars=["Label"], 
    value_vars=["None", "mild", "medium", "hot"], 
    var_name="Spice Level", 
    value_name="Frequency"
)
melted_df = melted_df[melted_df["Frequency"] == 1]

melted_df

# Create a grouped bar chart
fig = px.bar(
    melted_df,
    x="Label",
    color="Spice Level",
    barmode="group",
    title="Frequency of Food Types by Spice Level",
    labels={"Label": "Food Type", "Frequency": "Count"}
)

# Show the plot
fig.show()
```
